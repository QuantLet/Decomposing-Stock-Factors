```python
import matplotlib.pyplot as plt
from PyEMD import EMD
import numpy as np
from scipy.signal import argrelextrema

data = np.random.seed(100)
data = np.random.random(100) - 0.5
max_peaks = argrelextrema(data, np.greater)
min_peaks = argrelextrema(data, np.less)
plt.figure(figsize = (18,6), facecolor='none')
plt.plot(data)
plt.scatter(max_peaks, data[max_peaks], c='r')
plt.scatter(min_peaks, data[min_peaks], c='b')
plt.legend()
#plt.xlabel('time (s)')
#plt.ylabel('Amplitude')
plt.gca().set_facecolor('none')
plt.title("Find Local Extrema")

import scipy.interpolate as spi
#data = np.random.random(100)-0.5
index = list(range(len(data)))
max_peaks = list(argrelextrema(data, np.greater)[0])
min_peaks = list(argrelextrema(data, np.less)[0])
ipo3_max = spi.splrep(max_peaks, data[max_peaks],k=3)
iy3_max = spi.splev(index, ipo3_max)
ipo3_min = spi.splrep(min_peaks, data[min_peaks],k=3)
iy3_min = spi.splev(index, ipo3_min)
iy3_mean = (iy3_max+iy3_min)/2
plt.figure(figsize = (18,6), facecolor='none')
plt.plot(data[2:-3])
plt.plot(iy3_max[2:-3])
plt.plot(iy3_min[2:-3])
plt.plot(iy3_mean[2:-3])
plt.legend()
#plt.xlabel('time (s)')
#plt.ylabel('microvolts (uV)')
plt.gca().set_facecolor('none')
plt.title("Cubic Spline Interpolation")

plt.figure(figsize = (18,6), facecolor='none')
plt.plot(data[2:-3])
plt.plot(data[2:-3]-iy3_mean[2:-3])
plt.plot(iy3_mean[2:-3])
plt.legend()
#plt.xlabel('time (s)')
#plt.ylabel('microvolts (uV)')
plt.gca().set_facecolor('none')
plt.title("Define IMF1")
```
