```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import hilbert

# 20250514 WKH RT produce an Envelope via Hilbert transform

# Time vector
t = np.linspace(0, 1, 1000)
high_freq = 50  # Frequency in Hz
omega_high = 2 * np.pi * high_freq

# Amplitude modulation: envelope is a slow sine wave
amplitude_envelope = 1 + 0.5 * np.sin(2 * np.pi * 2 * t)

# Modulated signal: high-frequency sine modulated by slow envelope
x_modulated = amplitude_envelope * np.sin(omega_high * t)

# Compute analytic signal and Hilbert transform
analytic_signal_modulated = hilbert(x_modulated)
hilbert_transform_modulated = np.imag(analytic_signal_modulated)
envelope = np.abs(analytic_signal_modulated)

# Plot modulated signal and envelope
plt.figure(figsize=(10, 5))
plt.plot(t, x_modulated, label='Modulated Signal: A(t)·sin(ωt)', linestyle='--')
plt.plot(t, envelope, label='Envelope (|analytic signal|)', color='red')
plt.title('Envelope of a Modulated Signal via Hilbert Transform')
plt.xlabel('Time')
plt.ylabel('Amplitude')
#plt.legend()
#plt.grid(True)
plt.savefig("Envelope_Hilbert.png", transparent = True)
plt.tight_layout()
plt.show()
```
