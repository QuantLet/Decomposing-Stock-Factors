```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Parameter
n_folds = 10
samples_per_fold = 10
colors = {
    "train": "#4CAF50",   # grün
    "val": "#2196F3",     # blau
    "test": "#FFC107"     # gelb
}

fig, ax = plt.subplots(figsize=(10, 2.5), facecolor = "none")

# 10 CV-Folds
for i in range(n_folds):
    y = n_folds - i
    for j in range(n_folds):
        color = colors["val"] if j == i else colors["train"]
        rect = patches.Rectangle((j * samples_per_fold, y), samples_per_fold, 0.8, facecolor=color, edgecolor='k')
        ax.add_patch(rect)

# Labeling each fold
for i in range(n_folds):
    ax.text(-12, n_folds - i + 0.35, f"Fold {i+1}", fontsize=10, verticalalignment='center')

ax.set_facecolor("none")
ax.set_xlim(-20, n_folds * samples_per_fold + 5)
ax.set_ylim(0.5, n_folds + 1)
ax.set_xticks([])
ax.set_yticks([])
ax.set_title("10-Fold Cross-Validation", fontsize=12, weight='bold')

plt.tight_layout()
plt.show()
```
